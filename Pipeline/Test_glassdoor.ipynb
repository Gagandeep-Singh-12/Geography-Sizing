```python
import Glassdoor
import pandas as pd
```

```python
import us
def Glassdoor_func(Glassdoor_obj):
    Glassdoor_obj.Scrape()
    locations = {}

    if Glassdoor_obj.Status == "Location Found":
        c=[]
        for i in Glassdoor_obj.locations_data:
            city, country = i.split(", ")
            if len(country) == 2 and us.states.lookup(country) != None:
                if "us" in list(locations.keys()):
                    locations["us"].append(country)
                else:
                    locations["us"] = [country]
            else:
                code = False
                if code == False:
                    continue 
                locations[code[1]] = [city]
    return locations
```

```python
df = pd.read_csv("/home/celebal/Data/companies_data.csv")
for i in range(10):
    url = df["website"].iloc[i]
    cn = df["company_name"].iloc[i]
    obj = Glassdoor.Scraper(cn,url)
    print(Glassdoor_func(obj))

```
